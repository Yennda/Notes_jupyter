```python
def omg2lm(omg):
    return 2*m.pi*c.c/omg
def lm2omg(lm):
    return 2*m.pi*c.c/lm
```

```python
def e(omg, omgF, G):
    return 2 * (omg-omgF) / G

def sigma(e, q):
    return (e + q)**2/(e**2 + 1)

def G(lmF, lmd):
    return lm2omg(lmF - lmd/2) - lm2omg(lmF + lmd/2)
```

```python
L = 540e-9

lmF = 750e-9
lmd = 30e-9
omgF = lm2omg(lmF)

lms = np.arange(700, 800)*1e-9
omgs = lm2omg(lms)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
# axes.grid(linestyle='--')
axes.set_title('Fano')
axes.set_xlabel('$\lambda$ [nm]')
axes.set_ylabel('$\sigma$ [a. u.]')

# axes_rel = axes.twinx()
# axes_rel.grid(linestyle='--')
# axes_rel.set_ylabel('relative intensity')

axes.plot(lms, sigma(e(omg2lm(lms), omgF, G(lmF, 30e-9)), 1) , linewidth=2, label = 'G = 30e-9')
axes.plot(lms, sigma(e(omg2lm(lms), omgF, G(lmF, 10e-9)), 1) , linewidth=2, label = 'G = 10e-9')


    
    

lgd = fig.legend(loc='center left', bbox_to_anchor=(1.3, 0.5))

# fig.savefig(folder + name +'.png', dpi=300, bbox_extra_artists=(lgd,), bbox_inches='tight')    
```
