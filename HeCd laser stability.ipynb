```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as c
import math as m

red='#DD5544'
blue='#0284C0'
folder='C:/Users/Bukacek/OneDrive - UFE/thesis/diplomka/litografie/intensity_measurements/'
```

```python
file='PM2935003.dat'

f= open(folder+file,"r")
contents=f.readlines()

wavelength=[]
intensity=[]

for line in contents[13:-1]:
#     line_split=line[:-2].split('\t')
#     line_split[0]=line_split[0].replace(',', '.')
#     line_split[1]=line_split[1].replace(',', '.')

    intensity.append(float(line)*1e6)   


    
```
